```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier, RandomForestClassifier, BaggingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import VotingClassifier
```

```python
# Load the training and testing dataset and Hyperparameter tuned models dataframe
results_df = pd.read_csv('../datasets/raw_datasets/results_df.csv')
df_train_copy = pd.read_csv('../datasets/raw_datasets/df_train_copy_preprocessed.csv')
df_test = pd.read_csv('../datasets/raw_datasets/df_test_preprocessed.csv')
sub = pd.read_csv('../datasets/raw_datasets/sample_submission.csv')
```

```python
# Testing Features
features = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Group',
            'Cabin_number', 'HomePlanet_Earth', 'HomePlanet_Europa', 'HomePlanet_Mars',
            'CryoSleep_False', 'CryoSleep_True', 'Destination_55 Cancri e',
            'Destination_PSO J318.5-22', 'Destination_TRAPPIST-1e', 'VIP_False',
            'VIP_True', 'Cabin_deck_A', 'Cabin_deck_B', 'Cabin_deck_C',
            'Cabin_deck_D', 'Cabin_deck_E', 'Cabin_deck_F', 'Cabin_deck_G',
            'Cabin_deck_T', 'Cabin_side_P', 'Cabin_side_S']
```

```python
# Iterate over each row in results_df
for _, row in results_df.iterrows():
    
    # Convert string back to dictionary
    best_params = eval(row['Best Params'])

    # Dynamically create an instance of the model with the best parameters
    best_model = eval(row['Model'])(**best_params)

    # Fit the model with Preprocessed Training Data.
    best_model.fit(df_train_copy[features], df_train_copy["Transported"])
    
    # Make predictions
    predictions = best_model.predict(df_test[features])

    sub['Transported'] = predictions
    sub['Transported'] = sub['Transported'].replace({0:False, 1:True})

    # Save predictions to a CSV file
    submission_path = f"../datasets/submissions/{row['Model']}_submission.csv"
    sub.to_csv(submission_path, index=False)

    print(f"Predictions saved to {submission_path}")
```

```python
best_models = []
for _, row in results_df.iterrows():
    # Convert string back to dictionary
    best_params = eval(row['Best Params'])
    # Dynamically instantiate the model class
    best_model = eval(row['Model'])(**best_params)
    best_models.append((row['Model'], best_model))

# Define the voting classifiers
hard_voting_clf = VotingClassifier(estimators=best_models, voting='hard')
soft_voting_clf = VotingClassifier(estimators=best_models, voting='soft')

# Fit the voting classifiers with Preprocessed Training Data
hard_voting_clf.fit(df_train_copy[features], df_train_copy["Transported"])
soft_voting_clf.fit(df_train_copy[features], df_train_copy["Transported"])

# Make predictions on the Preprocessed Testing Data
hard_predictions = hard_voting_clf.predict(df_test[features])
soft_predictions = soft_voting_clf.predict(df_test[features])

# Convert predictions to the required format (e.g., boolean values)
sub['Transported'] = hard_predictions
sub['Transported'] = sub['Transported'].replace({0: False, 1: True})
# Save hard voting predictions to a CSV file
hard_submission_path = "../datasets/submissions/hard_voting_classifier_submission.csv"
sub.to_csv(hard_submission_path, index=False)
print(f"Hard Voting Predictions saved to {hard_submission_path}")

sub['Transported'] = soft_predictions
sub['Transported'] = sub['Transported'].replace({0: False, 1: True})
# Save soft voting predictions to a CSV file
soft_submission_path = "../datasets/submissions/soft_voting_classifier_submission.csv"
sub.to_csv(soft_submission_path, index=False)
print(f"Soft Voting Predictions saved to {soft_submission_path}")
```
